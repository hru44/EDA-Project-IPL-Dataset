```python
# IPL Finals venues and winners along with the number of wins.

matches_won_at_venue = data1.groupby(['venue','winner'])['id'].count().reset_index().head(50)
matches_won_matrix = matches_won_at_venue.pivot('venue','winner','id').fillna(0)

plt.figure(figsize = (12,8))
matches_won_matrix.plot(kind = 'bar', stacked = True, colormap = 'tab20')
plt.xlabel('venue')
plt.ylabel('Number of matches won')
plt.title('Number of matches won by each team at each venue')

plt.legend(title = 'winner', bbox_to_anchor = (1,1))
plt.show()
```

```python
# The man of the match for each winning team
man_of_the_match = data1.groupby(['winner','player_of_match'])['id'].count().reset_index()
a = man_of_the_match.pivot('winner','player_of_match','id').fillna(0)
plt.figure(figsize = (12,6))
sns.heatmap(a)
plt.title('The man of the match for each winning team')
plt.xlabel=('Player of the match')
plt.ylabel=('Winning team')
```

```python
# The toss winner, toss decision, winner in final matches.

toss_winners = data1.groupby(['toss_winner','toss_decision'])['winner'].count().reset_index()
a = toss_winners.pivot('toss_winner','toss_decision','winner').fillna(0)

plt.figure(figsize =(12,6))
a.plot(kind = 'bar', colormap='tab20')
plt.xlabel=('Toss winner')
plt.ylabel=('Toss decision')
plt.title('The Toss winner,toss decision,winner in final matches.')
```

```python
# The total runs by fours hit and the total number of fours hit by each team

fours = data[data['batsman_runs']==4]
a = fours.groupby('batting_team')['batsman_runs'].agg([('Total runs by four','sum'),('Fours','count')])
a
```

```python
plt.figure(figsize = (12,5))
a.plot(kind = 'bar', colormap = 'tab20')
plt.xlabel=('Batting_team')
plt.ylabel=('Fours')
plt.title('The total runs by fours hit by each team.')
```

Inferance of above graph

The Total number of fours hit by Mumbai Indians

```python
#The fours hit by players
batsman_four = data[data['batsman_runs']==4]
a = batsman_four.groupby('batsman')['batsman_runs'].agg([('four','count')]).reset_index()
a
```

```python
b = a.iloc[:10,:].plot('batsman','four',kind = 'bar', color = 'red')
plt.xlabel = ('Player name')
plt.ylabel = ('Number of fours')
plt.title('The fours hit by players')
plt.show()
```

Inferance of above graph

In chart, we show the most fours hit by A Mishra

```python
#The total runs by the sixes hit and the number of sixes hit by each team.

six_data = data[data['batsman_runs']==6]
a = six_data.groupby('batting_team')['batsman_runs'].agg([('Total runs by six','sum'),('six','count')])
a
```

```python
plt.figure(figsize=(12,8))
a.plot(kind='bar', colormap='tab20')
plt.xlabel=('batting team')
plt.ylabel=('Number of six')
plt.title('The total runs by six hit by each team')
```

Inferance of above graph

In this graph we show that most runs hitted by six royal challengers bangalore

```python
# The number of sixes hit in each season

sixes = data[data['batsman_runs']==6]
a = sixes.groupby('batsman')['batsman_runs'].agg([('six','count')]).reset_index()
a
```

```python
b = a.iloc[:10,:].plot('batsman','six', kind = 'bar', color = 'yellow')
plt.xlabel = ('Player name')
plt.ylabel = ('Number of six')
plt.title('The six hit by players ')
plt.show()
```

Inferance of above graph

The most six hit by A Symonds
